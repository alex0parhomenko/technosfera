```python
import sys
import re
import random
from operator import itemgetter
import urlparse
```

```python
def extract_features(INPUT_FILE_1, INPUT_FILE_2, OUTPUT_FILE):
    f_in1 = open(INPUT_FILE_1, "r")
    f_in2 = open(INPUT_FILE_2, "r")
    f_out = open(OUTPUT_FILE, "w")
    url_list = []
    max_url_cou = 1000
    for num, link in enumerate(f_in1):
        if (num > max_url_cou):
            break
        link = link.strip("\n")
        link = link.strip("/")
        link_arr = urlparse.urlparse(link).path.split("/")[1:len(link)]
        url_list.append(link_arr)
        
    for num, link in enumerate(f_in2):
        if (num > max_url_cou):
            break
        link = link.strip("\n")
        link = link.strip("/")
        link_arr = urlparse.urlparse(link).path.split("/")[1:len(link)]
        url_list.append(link_arr)
        
    print url_list
    
```

```python
extract_features("data/urls.lenta.general", "data/urls.lenta.examined", "out.txt")
```
